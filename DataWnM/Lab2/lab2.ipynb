```python
file = open("C:/Users/adivi/Documents/GitHub/practiceprograms/DataWnM/Lab1/transactions3.txt", 'r')
data = file.readlines()
file.close()

freq = {}

for line in data:
    items = line.split('-')[1].split(' ')
    for item in items:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

order = ['A', 'B', 'C', 'D', 'E']
for item in order:
    frequency = freq.get(item, 0)
    print(item+"-"+str(frequency))
```

```python
file = open("C:/Users/adivi/Documents/GitHub/practiceprograms/DataWnM/Lab1/transactions3.txt", 'r')
data = file.readlines()
file.close()

trans = {}

for line in data:
    tid, items = line.split('-')
    for item in items.split():
        if item in trans:
            trans[item].append(tid)
        else:
            trans[item] = [tid]

order = ['A', 'B', 'C', 'D', 'E']

newfile = open("transactions4.txt", 'w')
for item in order:
        transactions = ' '.join(trans.get(item, []))
        newfile.write(item+"-"+transactions+"\n")
newfile.close()

print("transactions4.txt Created")
```
