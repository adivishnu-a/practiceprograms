```python
import warnings
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

warnings.filterwarnings("ignore")

irisdata = load_iris()
dataset = pd.DataFrame(irisdata.data, columns=irisdata.feature_names)

k = 3  
kmeans = KMeans(n_clusters=k)

kmeans.fit(dataset)

labels = kmeans.labels_

dataset['Cluster'] = labels

cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=irisdata.feature_names)
print("Cluster Centers:")
print(cluster_centers)

plt.scatter(dataset['sepal length (cm)'], dataset['sepal width (cm)'], c=labels)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title(f'K-means Clustering (K={k})')
plt.show()
```
